### Sift the image

```python
shift_img = cv2.imread('dog.jpg')
shift_img = cv2.cvtColor(shift_img, cv2.COLOR_BGR2RGB)
plt.imshow(shift_img)
plt.show()  
```

```python
row, col, channel = shift_img.shape
print(row, col, channel)
```

```python
for i in range(row):
    for j in range(col):
        for k in range(channel):
            shift_img[i][(j+50)%col][k] = shift_img[i][j][k] 
plt.imshow(shift_img)
plt.show()
```

### increase the Brightness

```python
now_img = cv2.imread('dog.jpg')
now_img = cv2.cvtColor(now_img, cv2.COLOR_BGR2RGB)
plt.imshow(now_img)
plt.show()
```

```python
x,y,z = now_img.shape
print(x,y,z)
```

```python
for i in range(x):
    for j in range(y):
        for k in range(3):
            now_img[i][j][k] = 255 if now_img[i][j][k] + 50 > 255 else now_img[i][j][k] + 50

plt.imshow(now_img)
plt.show()
```
